## 1. Importar librerías

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

## 2. Cargar datos (Data Ingestion)

```python
# Cargar dataset
df = pd.read_csv("employee_churn_sample.csv")

# Exploración inicial (esto normalmente NO iría en un script productivo)
print("Primeras filas del dataset:")
print(df.head())

print("\nInformación del dataset:")
print(df.info())
```

## 3. Validación de datos

```python
# Validación simple de datos
print("Verificando valores nulos...")
print(df.isnull().sum())

assert df.isnull().sum().sum() == 0, "Existen valores nulos en el dataset"

print("Validación completada")
```

## 4. Preprocesamiento

```python
# Separar features y target
X = df.drop("churn", axis=1)
y = df["churn"]

# Escalado
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print("Preprocesamiento completado")
```

## 5. División del dataset

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("Tamaño entrenamiento:", X_train.shape)
print("Tamaño prueba:", X_test.shape)
```

## 6. Entrenamiento del modelo

```python
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)

print("Modelo entrenado correctamente")
```

## 7. Evaluación del modelo

```python
predictions = model.predict(X_test)

accuracy = accuracy_score(y_test, predictions)

print("Accuracy:", accuracy)

print("\nReporte de clasificación:")
print(classification_report(y_test, predictions))
```

## 8. Inferencia de ejemplo

```python
sample = X_test[:5]

pred = model.predict(sample)

print("Predicciones ejemplo:", pred)
```
